### 1. Which establishments have a hygiene score equal to 20?

```python
input_hygiene_score = int(input("Enter a hygiene score of interest: ")) # I created an input variable, but used 20 for this section of the challenge, as requested.

# Find the establishments with a hygiene score of 20
query1 = {"scores.Hygiene": input_hygiene_score}

# Use count_documents to display the number of documents in the result
hygiene_20_count = establishments.count_documents(query1)
print(f'There are {hygiene_20_count} establishments with a hygiene score of {input_hygiene_score}')

# Display the first document in the results using pprint
hygiene_20_doc = establishments.find_one(query1)
print(f"First establishment with a hygiene score of {input_hygiene_score}: ")
pprint(hygiene_20_doc)
```

```python
# Retrieve all documents in the collection for establishments with a hygiene score of 20
all_hygiene_20_docs = establishments.find(query1)

# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.DataFrame(list(all_hygiene_20_docs))

# Display the number of rows in the DataFrame
print(f'The number of rows in hygiene_20_df: {len(hygiene_20_df)}\n')

# Display the first 10 rows of the DataFrame
print("Herewith the first 10 rows of the DataFrame: ")
hygiene_20_df.head(10)
```

```python
# Note: four columns above are nested structures, so we need to flatten them.

from pandas import json_normalize

# Ensure the links field is properly formatted
def flatten_links(links):
    if isinstance(links, list) and links:
        return links[0]
    else:
        return {}

hygiene_20_df['links'] = hygiene_20_df['links'].apply(flatten_links)

# Flatten the nested structures using json_normalize
scores_df = json_normalize(hygiene_20_df['scores'])
geocode_df = json_normalize(hygiene_20_df['geocode'])
meta_df = json_normalize(hygiene_20_df['meta'])
links_df = json_normalize(hygiene_20_df['links'])

# Drop the nested structures if they exist
columns_to_drop = ['scores', 'geocode', 'meta', 'links']
existing_columns_to_drop = [col for col in columns_to_drop if col in hygiene_20_df.columns]
hygiene_20_df = hygiene_20_df.drop(columns=existing_columns_to_drop)

# Merge the flattened dataframes with the original dataframe
hygiene_20_df = pd.concat([hygiene_20_df, scores_df, geocode_df, meta_df, links_df], axis=1)

# Display the number of rows in the DataFrame
print(f'The number of rows in the DataFrame: {len(hygiene_20_df)}\n')

# Display the first 10 rows of the DataFrame
print("Here are the first 10 rows of the DataFrame: ")
hygiene_20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = {"LocalAuthorityName":{"$regex": "London", "$options": "i"}, "RatingValue": {"$gte": 4}}

# Use count_documents to display the number of documents in the result
rating_4_count = establishments.count_documents(query2)
print(f"Number of establishments in London with a RatingValue greater than or equal to 4: {rating_4_count}\n")

# Display the first document in the results using pprint
first_rating_4_doc = establishments.find_one(query2)
print('The first establishment in London with a RatingValue greater than or equal to 4: ')
pprint(first_rating_4_doc)
```

```python
# Retrieve all documents in the collection for establishments with a hygiene score of 20
all_rating_4_docs = establishments.find(query2)

# Convert the result to a Pandas DataFrame
rating_4_df = pd.DataFrame(list(all_rating_4_docs))

# Display the number of rows in the DataFrame
print(f'The number of rows in the DataFrame: {len(rating_4_df)}\n')

# Display the first 10 rows of the DataFrame
print("Herewith the first 10 rows of the DataFrame: ")
rating_4_df.head(10)
```

```python
# Ensure the links field is properly formatted
def flatten_links(links):
    if isinstance(links, list) and links:
        return links[0]
    else:
        return {}

rating_4_df['links'] = rating_4_df['links'].apply(flatten_links)

# Flatten the nested structures using json_normalize
scores_df = json_normalize(rating_4_df['scores'])
geocode_df = json_normalize(rating_4_df['geocode'])
meta_df = json_normalize(rating_4_df['meta'])
links_df = json_normalize(rating_4_df['links'])

# Drop the nested structures if they exist
columns_to_drop = ['scores', 'geocode', 'meta', 'links']
existing_columns_to_drop = [col for col in columns_to_drop if col in rating_4_df.columns]
rating_4_df = rating_4_df.drop(columns=existing_columns_to_drop)

# Merge the flattened dataframes with the original dataframe
rating_4_df = pd.concat([rating_4_df, scores_df, geocode_df, meta_df, links_df], axis=1)

# Display the number of rows in the DataFrame
print(f'The number of rows in the DataFrame: {len(rating_4_df)}\n')

# Display the first 10 rows of the DataFrame
print("Here are the first 10 rows of the DataFrame: ")
rating_4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Retrieve the coords for Penang Flavours
penang_flavours = establishments.find_one({"BusinessName": "Penang Flavours"})
latitude = float(penang_flavours['geocode']['latitude'])
longitude = float(penang_flavours['geocode']['longitude'])

# Rating value must equal 5

# Define the search parameters
degree_search = 0.01
lat_min = latitude - degree_search
lat_max = latitude + degree_search
long_min = longitude - degree_search
long_max = longitude + degree_search

# Define the query
query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": lat_min, "$lte": lat_max},
    "geocode.longitude": {"$gte": long_min, "$lte": long_max}
}

# Sort by hygiene score
sort = [("scores.Hygiene", 1)]

# Limit to top 5 establishments
limit = 5

# Retrieve the top 5 establishments
top_5 = establishments.find(query).sort(sort).limit(limit)


```

```python
# Convert the result to a Pandas DataFrame
top_5_list = list(top_5)
top_5_df = pd.DataFrame(top_5_list)

# Print the DataFrame
print("Top 5 establishments with RatingValue of 5, sorted by lowest hygiene score, nearest to 'Penang Flavours':")
top_5_df.head(6)

# Print the list of top 5 BusinessNames
top_5_business_names = top_5_df['BusinessName'].tolist()
print("List of the top 5 BusinessNames:")
print(top_5_business_names)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
pipeline = [
    # 1. Matches establishments with a hygiene score of 0
    {"$match": {"scores.Hygiene": 0}},
    
    # 2. Groups the matches by Local Authority
    {"$group": {
        "_id": "$LocalAuthorityName",
        "count": {"$sum": 1}
    }},
    
    # 3. Sorts the matches from highest to lowest
    {"$sort": {"count": -1}}
]

# Execute the aggregation pipeline
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of Local Authorities with establishments having a hygiene score of 0: {len(results)}\n")

# Print the first 10 results
print("Top 10 Local Authorities with establishments having a hygiene score of 0:")
pprint(results[:10])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results)

# Rename the columns for better readability
results_df.rename(columns={"_id": "LocalAuthorityName", "count": "EstablishmentCount"}, inplace=True)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(results_df)}\n")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
results_df.head(10)
```
